```python
import numpy as np
```

```python

print("="*50)
print("üßÆNumpy CalculatorüßÆ".center(50))
print("="*50)
print("\nChoose an option:")
print("1. Arithmetic Operations")
print("2. Statistical Operations")
print("3. Array Mnipulation")
print("4. Exit")

choice=int(input("\nEnter Your Choice (1-4):"))

#-------------------------------------------------------------------------------


class calculator:

  @staticmethod
  def Arithmetic_operation():
    print("\n"+"-"*50)
    print("BASIC ARITHMETIC OPERATIONS:".center(50))
    print("-"*50)

    arr1=np.array(input("\nEnter values by spaces:").split()).astype(float)
    arr2=np.array(input("Enter values by spaces:").split()).astype(float)


    if arr1.shape == arr2.shape:

      print("\nADDITION:",np.add(arr1,arr2))
      print("SUBTRACTION:",np.subtract(arr1,arr2))
      print("MULTIPLICATION:",np.multiply(arr1,arr2))
      print("DIVISION:",np.divide(arr1,arr2))
      print("POWER:",np.power(arr1,arr2))

      action=input("Press Enter to continue:")

      if action =="":
        choice=int(input("\nEnter Your Choice (1-4):"))

        if choice == 2:
          calculator.Statistical_operation()
        elif choice == 3:
          calculator.Array_manipulation()
        elif choice == 4:
          print("Exit")

      else:
        if action=="NO" or action=="no" or action=="No":
          print("Exit")



    else:
      print("\nArray shape should be same")
      arr1=np.array(input("\nEnter values Again seperated by spaces:").split()).astype(float)
      arr2=np.array(input("Enter values Again seperated by spaces:").split()).astype(float)

      print("\nADDITION:",np.add(arr1,arr2))
      print("SUBTRACTION:",np.subtract(arr1,arr2))
      print("MULTIPLICATION:",np.multiply(arr1,arr2))
      print("DIVISION:",np.divide(arr1,arr2))
      print("POWER:",np.power(arr1,arr2))

      action=input("Press Enter to continue otherwise Press No:")

      if action =="":
        choice=int(input("\nEnter Your Choice (1-4):"))

        if choice == 2:
          calculator.Statistical_operation()
        elif choice == 3:
          calculator.Array_manipulation()
        elif choice == 4:
          print("Exit")

      else:
        if action=="NO" or action=="no" or action=="No":
          print("Exit")





#-------------------------------------------------------------------------------

  @staticmethod
  def Statistical_operation():
    print("\n"+"-"*50)
    print("STATISTICAL OPERATIONS:".center(50))
    print("-"*50)

    arr3=np.array(input("Enter the values seperated by spaces:").split()).astype(float)

    try:

      print("Mean:",np.mean(arr3))
      print("Median:",np.median(arr3))
      print("Standard Deviation:",np.std(arr3))
      print("Variance:",np.var(arr3))
      print("Minimum:",np.min(arr3))
      print("Maximum:",np.max(arr3))
      print("Sum:",np.sum(arr3))
      print("Product:",np.prod(arr3))
      print("Shape:",arr3.shape)
      print("Dimention:",arr3.ndim)
      print("Size:",arr3.size)
      print("Datatype:",arr3.dtype)

      action=input("Press Enter to continue otherwise Press No:")

      if action =="":
        choice=int(input("\nEnter Your Choice (1-4):"))

        if choice == 1:
          calculator.Arithmetic_operation()
        elif choice == 3:
          calculator.Array_manipulation()
        elif choice == 4:
          print("Exit")

      else:
        if action=="NO" or action=="no" or action=="No":
          print("Exit")


    except Exception as e:
      print(f"An error occured:{e}")


#-------------------------------------------------------------------------------

  @staticmethod
  def Array_manipulation():
    print("\n"+"-"*50)
    print("ARRAY MANIPULATION:".center(50))
    print("-"*50)



    arr4=np.array(input("Enter values separated by spaces:").split()).astype(float)

    try:

        print("Origianl array:",arr4)
        print("Sorted array:",np.sort(arr4))
        print("Reversed array:",np.sort(arr4)[::-1])
        print("Unique values:",np.unique(arr4))
        print("Accessing first ele:",arr4[0])
        print("Accessing last ele:",arr4[-1])

        rows=int(input("\nEnter the row value for reshape:"))
        columns=int(input("Enter the column value for reshape:"))
        elements=int(input("Enter the element value for reshape:"))

        if (rows*columns*elements) == arr4.size:
          print("3D Array:",arr4.reshape(rows,columns,elements))

        elif (rows*columns) == arr4.size:
          print("2D Array:",arr4.reshape(rows,columns))

        else:
          print("Eror:Invalid Array's reshaping rows and columns")


        action=input("Press Enter to continue:")

        if action =="":
          choice=int(input("\nEnter Your Choice (1-4):"))

          if choice == 2:
            calculator.Statistical_operation()
          elif choice == 1:
            calculator.Arithmetic_operation()
          elif choice == 4:
            print("Exit")

        else:
          if action=="NO" or action=="no" or action=="No":
            print("Exit")

    except Exception as e:
      print(f"An Error occured:{e}")




if choice == 1:
  calculator.Arithmetic_operation()
elif choice == 2:
  calculator.Statistical_operation()
elif choice == 3:
  calculator.Array_manipulation()
elif choice == 4:
  print("Exit")






```
